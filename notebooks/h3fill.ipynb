# Filling with h3 hexagons

```python
import tobler
from shapely.geometry import Polygon
import geopandas
import libpysal
%load_ext watermark
%watermark -v -a "serge rey" -d -u -p tobler,libpysal,geopandas
```

```python
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (16, 9)
```

```python
sids2 = libpysal.examples.load_example('SIDS2')
gdf = geopandas.read_file(sids2.get_path('sids2.shp'))
```

```python
gdf.plot()
```

```python
gdf.crs
```

```python
h = tobler.util.h3fy(gdf, resolution=6)
```

```python
h.plot()
```

```python
gdfe = gdf.explode(index_parts=True)
gdfe['state']='nc'
gdfe.dissolve(by='state').plot()
```

```python
f, ax = plt.subplots(1,1, figsize=(16,9))
h.plot(ax=ax)
gdfe.dissolve(by='state').plot(alpha=0.7, facecolor='none', ax=ax);
```

We see the unfilled spaces near the boundary. This is the issue raised in [#146](https://github.com/pysal/tobler/issues/146).

## Using `buffer` and `distance` to ensure complete fill

```python
gdfutm = gdf.to_crs(gdf.estimate_utm_crs())
```

```python
gdfutm.plot(figsize=(16,9));
```

```python
h = tobler.util.h3fy(gdfutm, resolution=6)
```

```python
hb = tobler.util.h3fy(gdfutm, resolution=6, buffer=True) # this will buffer but with a distance of 0
```

```python
hb.plot();
```

```python
hbc = tobler.util.h3fy(gdfutm, resolution=6, buffer=True, clip=True) # this will clip but buffer distance is 0 so default behavior
```

```python
hbc.plot();
```

```python
gdfutm['state']='nc'
f, ax = plt.subplots(1,1, figsize=(16,9))
hb.plot(ax=ax)
gdfutm.dissolve(by='state').plot(alpha=0.7, facecolor='none', ax=ax);
```

```python
gdfutm['state']='nc'
f, ax = plt.subplots(1,1, figsize=(16,9))
hbc.plot(ax=ax)
gdfutm.dissolve(by='state').plot(alpha=0.7, facecolor='none', ax=ax);
```

Using buffer without a non-zero distance maintains the previous implementation.

If we specify a distance for the buffer in the proper CRS units, we can correct the empty spaces:

```python
hbc = tobler.util.h3fy(gdfutm, resolution=6, buffer=True, clip=True, distance=10000);
```

```python
gdfutm['state']='nc'
f, ax = plt.subplots(1,1, figsize=(16,9))
hbc.plot(ax=ax)
gdfutm.dissolve(by='state').plot(alpha=0.7, facecolor='none', ax=ax);
```

## If needed, clip can be turned off

```python
hbc = tobler.util.h3fy(gdfutm, resolution=6, buffer=True, clip=False, distance=10000);
```

```python
gdfutm['state']='nc'
f, ax = plt.subplots(1,1, figsize=(16,9))
hbc.plot(ax=ax)
gdfutm.dissolve(by='state').plot(alpha=0.7, facecolor='none', ax=ax);
```
